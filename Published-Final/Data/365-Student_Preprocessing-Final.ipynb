# Adding Course Data as columns

```python
le=pd.read_csv('365_student_learning.csv')
```

```python
le
```

```python
le.drop(columns=['minutes_watched','date_watched'],inplace=True)
le
```

```python
le.drop_duplicates(inplace=True)
```

```python
le
```

```python
np.sort(le['course_id'].unique())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([('ohenc', OneHotEncoder(categories=[np.sort(le['course_id'].unique())],sparse=False),  ['course_id'])])
x=ct.fit_transform(le)
```

```python
x.shape
```

```python
# le['course_id']
cols=np.sort(le['course_id'].unique())
cols
```

```python
df_courses=pd.DataFrame(x,columns=cols)
df_courses
```

```python
df_courses.iloc[:10,:]
```

```python
le
```

```python
le.reset_index(inplace=True)
le
```

```python
le['student_id']
```

```python
df_courses
```

```python
df_courses['student_id']=le['student_id']
df_courses
```

```python
df_courses=df_courses.groupby('student_id').sum()
```

```python
df_courses
```

```python
data
```

```python
df_student_features_final=data.merge(df_courses,on='student_id',how='left')
```
